```python
import pandas as pd
import re 

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException # mudar de paginas
```

```python
service = Service()

options= webdriver.ChromeOptions()

driver =webdriver.Chrome(service=service, options=options)
```

```python
max_pages = 5
page = 1

titleList =[]
stockList=[]
priceList=[]
genderList=[]

while page <= max_pages:
    url = f'https://books.toscrape.com/catalogue/page-{page}.html'
    driver.get(url)

    elementos = driver.find_elements(By.CSS_SELECTOR, "ol.row a")

    for i in range(0,len(elementos),2):
        elemento = elementos[i]
        elemento.click()

        title = driver.find_element(By.TAG_NAME, 'h1').text
        titleList.append(title)

        texto_stock = driver.find_element(By.CLASS_NAME, 'instock').text 
        numero_stock = int(re.search(r'\d+', texto_stock).group())
        stockList.append(numero_stock)

        texto_price =driver.find_element(By.CLASS_NAME, 'price_color').text 
        numero_price = int(re.search(r'\d+', texto_price).group())
        priceList.append(numero_price)

        links = driver.find_elements(By.CSS_SELECTOR, "ul a") # home | books | poetry
        gender=links[2].text
        genderList.append(gender)

        driver.back()

    page += 1

driver.close()

```

```python
df= pd.DataFrame({'Title':titleList, 'Stock':stockList, 'Price':priceList, 'Gender':genderList})
```

```python
df.sort_values(by = 'Title', inplace=True)
```

```python
df['In_stock'] = df['Stock'].apply(lambda x: 1 if x > 0 else 0)
```
